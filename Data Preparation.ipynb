```python
import numpy as np
import pandas as pd
```

## Read the CSV

```python
df = pd.read_csv('flipkart_laptop.csv')
df.head()
```

### Separate the Brand Name from name

```python
df['name'].str.split()[0]
```

```python
df['Brand'] = np.nan
for i in range(df.shape[0]):
    df['Brand'][i] = df['name'][i].split()[0]
    
df.head()
```

### Look for missing Values

```python
df.isnull().sum()
```

### Separate Ratings and No. of Reviews

```python
df.num_ratings[0]
```

```python
df.num_ratings[0].replace('\xa0','').split('&')
```

```python
df['num_reviews'] = np.nan
for i in range(df.shape[0]):
    if(not pd.isnull(df['num_ratings'][i])):
        rr = df['num_ratings'][i].replace('\xa0','').split('&')
        rat_temp = rr[0].split()[0]
        rev_temp = rr[1].split()[0]

        if (',' in rat_temp):
            df['num_ratings'][i] = int(rr[0].split()[0].replace(',',''))
        else:
            df['num_ratings'][i] = int(rr[0].split()[0])

        if (',' in rev_temp):
            df['num_reviews'][i] = int(rr[1].split()[0].replace(',',''))
        else:
            df['num_reviews'][i] = int(rr[1].split()[0])
    
        

df.head()
```

```python
df.dtypes
```

```python
df['num_ratings'] = pd.to_numeric(df['num_ratings'] , errors = 'coerce')
```

```python
df.dtypes
```

### Clean RAM Variable and extract ram size and ram type

```python
df['ram'][0].split()
```

```python
df['ram_type'] = np.nan
for i in range(df.shape[0]):
    ram_split = df['ram'][i].split()
    df['ram'][i] = ram_split[0]
    df['ram_type'][i] = ram_split[2]
```

```python
df.head()
```

### Clean Price Column

```python
for i in range(df.shape[0]):
    df['price'][i] = df['price'][i].replace('₹','').replace(',','')
    
df.head()
```

### Clean Original Price Column

```python
for i in range(df.shape[0]):
    if(not pd.isnull(df['org_price'][i])):
        df['org_price'][i] = df['org_price'][i].replace('₹','').replace(',','')
    else:
         df['org_price'][i] = df['org_price'][i]
    
df.head()
```

### Convert Price, org_price, ram and num_ratings to numeric

```python
df['price'] = pd.to_numeric(df['price'] , errors = 'coerce')
df['org_price'] = pd.to_numeric(df['org_price'] , errors = 'coerce')
df['ram'] = pd.to_numeric(df['ram'] , errors = 'coerce')
```
